```python
import zipfile
import requests
from io import BytesIO
import pandas as pd
```

```python
dataZipURL = "http://archive.ics.uci.edu/ml/machine-learning-databases/00222/bank-additional.zip"
csvFile = "bank-additional/bank-additional-full.csv"
r = requests.get(dataZipURL)
z = zipfile.ZipFile(BytesIO(r.content))
df = pd.read_csv(z.open(csvFile), sep=";")
print("Dataset size: %s" % str(df.shape))
df.head()
```

```python
import sys
from importlib import reload

customPackages = ["visualizations"]
for pkg in customPackages:
    reload(sys.modules[pkg]) if pkg in sys.modules else __import__(pkg)
vis = sys.modules["visualizations"].Visualizations()
```

```python
vis.drawKDEPlot(data=df, variable="cons.price.idx", title="Consumer Price Index KDE Plot")
```

```python
variables = list(df.columns[df.dtypes != object])
vis.drawMultipleKDEPlots(data=df, variables=variables, figSize=(18,12), title="KDE Plots", showGrid=True)
```
